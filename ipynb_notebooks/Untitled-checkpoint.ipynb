```python

import sys

sys.path.append('../src/')

import numpy as np
import matplotlib.pyplot as plt

import os
print(os.getcwd())
from main import *

# X = np.array(
#         [
#             [-3, 8],
#             [-1, -4],
#             [5, 5],
#             [8, 2],
#             [-7, -7],
#             [7, -10],
#             [-1, 4],
#             [-6, -10],
#             [-8, 9],
#             [-1, 3],
#         ]
# )
X = np.array(
        [
            [-2, 3],
            [0, 1],
            [4, -2],
            [6, 1],
            [-5, -4],
            [7, -6],
            [-3, 2],
            [-4, -3],
            [-6, 5],
            [2, -1],
        ]
)

y = np.array(
        [[5], [-2], [8], [15], [-9], [18], [2], [-7], [12], [6]]
)

n = X.shape[1]
weights = np.zeros(n + 1)
cost = 0


weights, cost = compute_linear_regression(X, y, 0.001, 10000)
print("Weights:", weights)
print("Cost:", cost)

# Extract the first feature (X-axis values)
x_values = X[:, 0]

# Generate the fitted line using: y = bias + slope * x
x_fit = np.linspace(x_values.min(), x_values.max(), 100)  # Generate smooth x values
y_fit = weights[0] + weights[1] * x_fit  # Compute corresponding y values

# Plot data points
plt.scatter(x_values, y, color='blue', label='Actual Data')

# Plot regression line
plt.plot(x_fit, y_fit, color='red', label='Best Fit Line')

# Labels and legend
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Linear Regression - Best Fit Line")
plt.legend()
plt.grid()

# Show plot
plt.show()

```
