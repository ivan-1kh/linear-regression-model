```python

import sys

sys.path.append('../src/')

import numpy as np
import matplotlib.pyplot as plt

import os
print(os.getcwd())
from main import *

X, y = generate_random_linear_dataset()

# Initialize weights and cost to store output
n = X.shape[1]
weights = np.zeros(n+1)
cost = 0

weights, cost = compute_linear_regression(X, y, 0.01, 10000)
print("Weights:", weights)
print("Cost:", cost)
compute_r_squared(X, y, weights) #Checks performance of the model


# Calculate predictions for the original X data
predictions = np.dot(X, weights[1:]) + weights[0]

# Extract the first feature (X-axis values)
x_values = X[:, 0]

# Generate the fitted line using: y = bias + slope * x
x_fit = np.linspace(x_values.min(), x_values.max(), 100)  # Generate smooth x values
y_fit = weights[0] + weights[1] * x_fit  # Compute corresponding y values


# Plot data points
plt.scatter(x_values, y, color='blue', label='Data Points')

# Plot predicted data points
plt.scatter(x_values, predictions, color='green', label='Predictions')

# Plot regression line
plt.plot(x_fit, y_fit, color='red', label='Best Fit Line')

# Labels and legend
plt.xlabel("Feature 1 values")
plt.ylabel("Y values")
plt.title("Linear Regression - Actual Data, Predictions and Best Line Fit")
plt.legend()
plt.grid(True)

# Show plot
plt.show()

```
